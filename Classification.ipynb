## **DATA PREPROCESSING**

### Data Cleaning

```python
# Check for missing values
print(data.isnull().sum())
```

Handling Missing values

```python
#Fill the missing values for numerical terms with mean
data["ApplicantIncome"] = data["ApplicantIncome"].fillna(data["ApplicantIncome"].mean())
data["CoapplicantIncome"] = data["CoapplicantIncome"].fillna(data["CoapplicantIncome"].mean())
data["LoanAmount"] = data["LoanAmount"].fillna(data["LoanAmount"].mean())
data["Loan_Amount_Term"] = data["Loan_Amount_Term"].fillna(data["Loan_Amount_Term"].mean())
data["Credit_History"] = data["Credit_History"].fillna(data["Credit_History"].mean())
```

```python
data['Gender'].mode()
```

```python
data['Gender'].mode()[0]
```

```python
#Fill the missing values for categarical terms with mode
data["Gender"] = data["Gender"].fillna(data["Gender"].mode()[0])
data["Married"] = data["Married"].fillna(data["Married"].mode()[0])
data["Dependents"] = data["Dependents"].fillna(data["Dependents"].mode()[0])
data["Education"] = data["Education"].fillna(data["Education"].mode()[0])
data["Self_Employed"] = data["Self_Employed"].fillna(data["Self_Employed"].mode()[0])
```

```python
#Check again for null values
data.isnull().sum()
```

### Descriptive Analysis








```python
data.describe()
```

###Predictive Analysis




```python
cols = ['Loan_ID']
data = data.drop(columns=cols, axis=1)
data.head()
```

Label Encoding

```python
#convert all categorical data into numeric form to make machine-readable
from sklearn.preprocessing import LabelEncoder

# Define the categorical features
categorical_features = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status', 'Dependents']

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Encode each categorical feature separately
for feature in categorical_features:
  data[feature] = encoder.fit_transform(data[feature])
```

Train Test Split

```python
x = data.drop(columns=['Loan_Status'], axis=1)
y = data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=50)
model.fit(x_train, y_train) #train
```

```python
pred = model.predict(x_test)
pred #predicted data from the model
```

```python
y_test #real data
```

Support Vector Classifier

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train, y_train) #train
```

```python
predi = model_svc.predict(x_test)
predi #predicted data from the model
```

```python
y_test
```

## **Evaluation and Interpretation**

### Results of Random Forest

```python
#check accuracy of the model
from sklearn .metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
#check the performance of the model
from sklearn .metrics import confusion_matrix
confusion_matrix (y_test, pred)
```

79 data is predicted correctly.17 data is predicted incorrectly.

Results for SVC

```python
from sklearn .metrics import accuracy_score
accuracy_score(y_test, predi)
```

```python
from sklearn .metrics import confusion_matrix
confusion_matrix (y_test, predi)
```

67 data is predicted correctly.29 data is predicted incorrectly.

### Validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, x, y, cv=5).mean() #cv=5 -- no of sets that you need to devide the data set
```
